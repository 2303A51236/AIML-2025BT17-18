```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
```

```python
dataset = pd.read_csv("loan_data.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
d=dataset.describe(percentiles=[.25,.5,.75,.9])
d
```

```python
print(dataset.columns)
```

```python
sns.lmplot(x='fico', y='int.rate', hue='not.fully.paid', col='credit.policy', data=dataset, height=6)
plt.show()
```

```python
dataset = pd.get_dummies(dataset, columns=['purpose'], drop_first=True)
```

```python
X = dataset.drop('not.fully.paid', axis=1)
Y = dataset['not.fully.paid']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
random = RandomForestClassifier(n_estimators=100)
random.fit(X_train, Y_train)
```

```python
predictions = random.predict(X_test)
```

```python
cm = confusion_matrix(Y_test, predictions)
print("Confusion Matrix:")
print(cm)
```

```python
cr = classification_report(Y_test, predictions)
print("Classification Report:")
print(cr)
```
